## Задача 1

**Написать функцию нейтрализации. На вход подается вектор произвольной длины, на выходе нейтрализованный вектор.**

```python
def neutralize(df: pd.DataFrame) -> pd.DataFrame:
    return df -  df.sum(axis=1) / df.shape[1]

def test_neutrality(df: pd.DataFrame) -> None:
    try:
        assert abs(df.sum().sum()) <= 1e-12
        print("Is Neutral")
    except Exception:
        print("Is not Neutral")
        
```

```python
test_neutrality(close_df)
test_neutrality(neutralize(close_df))
```

## Задача 2

**Написать функцию нормализации. На вход подается вектор произвольной длины, на выходе нормализованный вектор.**

```python
def normalize(df: pd.DataFrame) -> pd.DataFrame:
    return (df.T / df.abs().sum(axis=1)).T

def test_normality(df: pd.DataFrame) -> None:
    try:
        assert abs(df.abs().sum().sum() /df.shape[0] - 1.0) <= 1e-12
        print("Is Normal")
    except Exception:
        print("Is not Normal")
        
```

```python
test_normality(close_df)
test_normality(normalize(close_df))
```
